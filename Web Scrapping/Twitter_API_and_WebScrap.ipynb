# Getting Omicron tagged tweets

```python
query2="#recession"
data = {
    'text':[],
    'author_id':[],
    'created_at':[]
}
for i in tweepy.Paginator(client.search_recent_tweets,query2,
                        tweet_fields=['author_id','created_at'],
                           max_results=100).flatten(1000):
    data['text'].append(i.text)
    data['author_id'].append(i.author_id)
    data['created_at'].append(i.created_at)
```

```python
data
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
```

```python
len(df["author_id"])
```

```python
df.to_csv("tweets.csv",index=False)
```

```python
def extract_categories(baseurl):
    categories={}
    r=requests.get(baseurl)
    soup=BeautifulSoup(r.content)
    cl=soup.find('ul',class_='nav-list').ul.find_all('li')
    for i in cl:
        categories.update({i.text.strip():baseurl + i.a['href']})
    return categories
```
