```python
# finding the day in the week where most shows/movies have been released
```

```python
newdf["date_added"]=pd.to_datetime(newdf["date_added"])
newdf["day_of_week"]=newdf["date_added"].dt.dayofweek
```

```python
# Convert day_of_week into day "text"
```

```python
newdf.isnull().sum()
```

```python
# Function to convert day to text using dict
def convertday(x):
    if x==-1:
        return None
    ans={0:"Mon",1:"Tue",2:"Wed",3:"Thu",4:"Fri",5:"Sat",6:"Sun"}
    return ans[int(x)]
```

```python
newdf['day_of_week'] = newdf['day_of_week'].fillna(-1)
```

```python
newdf["day_of_week"]=newdf["day_of_week"].apply(convertday)
```

```python
newdf["day_of_week"].value_counts()
```

```python
# Graph4 : Day in which most content is released in a week
newdf["day_of_week"].value_counts().plot(kind="bar")
plt.show()
```

```python
# As per the bar graph4 we can easliy say that most of the content that is released is on Friday as it is a last working day and people 
# can binge watch the content :-D
```

```python
newdf.isnull().sum()
```

```python
newdf["date_added"]=pd.to_datetime(newdf["date_added"])
newdf["month_added"]=newdf["date_added"].dt.month
```

```python
# Function to convert month to text using dict
def convertmonth(x):
    if x==-1:
        return None
    ans={1:"Jan",2:"Feb",3:"Mar",4:"April",5:"May",6:"June",7:"July",8:"Aug",9:"Sept",10:"Oct",11:"Nov",12:"Dec"}
    return ans[int(x)]
```

```python
newdf['month_added'] = newdf['month_added'].fillna(-1)
newdf["month_added"]=newdf["month_added"].apply(convertmonth)
```

```python
newdf["month_added"].value_counts()
```

```python
# Graph5 : Month in which most content is released

graph5=newdf["month_added"].value_counts().sort_index().plot(kind="bar")
for i in graph5.containers:
    graph5.bar_label(i)

plt.show()
```

```python
# As per Graph5 we can see that "July" is the month where most content is released
```
